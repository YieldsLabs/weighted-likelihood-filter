```python
def rk4_step(y, i, dt, f):
    h = dt
    t = dt * i
    k1 = h * f(y, t)
    k2 = h * f(y + k1 / 2, dt * i + h / 2)
    k3 = h * f(y + k2 / 2, t + h / 2)
    k4 = h * f(y + k3, t + h)

    y_next = y + 1 / 6 * (k1 + 2 * k2 + 2 * k3 + k4)
    return y_next

    

@partial(jax.jit, static_argnames=("f",))
def rk4(ys, dt, N, f):
    """
    Based on
    https://colab.research.google.com
    github/google/jax/blob/master/cloud_tpu_colabs/Lorentz_ODE_Solver
    """
    @jax.jit
    def step(i, ys):
        ysi = rk4_step(ys[i - 1], i, dt, f)
        return ys.at[i].set(ysi)
    return jax.lax.fori_loop(1, N, step, ys)
```

## 3D Lorenz

```python
sigma = 10.
beta = 8./3
rho = 28.


def f(state, t):
    x, y, z = state
    xdot = sigma * (y - x)
    ydot = x * (rho - z) - y
    zdot = x * y - beta * z
    return jnp.array([xdot, ydot, zdot])

y0 = jnp.ones(3)
N = 5_000
ys = jnp.zeros((N,) + y0.shape)
ys = ys.at[0].set(y0)
res = rk4(ys, 0.004, N, f)
```

# Modified Lorenz 96 model
See https://en.wikipedia.org/wiki/Lorenz_96_model


$$
    \dot{\bf x}_k = \Big({\bf x}_{k+1} - {\bf x}_{k-2}\Big) {\bf x}_{j-1} - {\bf x}(j) + {F}_k
$$
with 
* $F_k \sim {\cal N}(8, 1)$

```python
D = 40
F = 8.0
ixs = jnp.arange(D)

key = jax.random.PRNGKey(31415)
key_init, key_sim, key_eval = jax.random.split(key, 3)
x0 = 8.0 + jax.random.normal(key_init, (D,))

key_state, key_measurement = jax.random.split(key_sim)
```

```python
@partial(jax.vmap, in_axes=(None, 0))
def fcoord(x, k):
    xdot = (x[(k + 1) % D] - x[k - 2]) * x[k - 1] - x[k] + F
    return xdot

def f(x, t, *args):
    keyt = jax.random.fold_in(key_state, t)
    F = jax.random.normal(keyt, shape=(D,)) * jnp.sqrt(0.05)
    xdot = fcoord(x, ixs) + F + 8.0
    return xdot
```

```python
snapshots = 10
N = 60
dt = 0.05
xs = jnp.zeros((N,) + x0.shape)
xs = xs.at[0].set(x0)
xs = rk4(xs, dt, N, f)
ys = xs + jax.random.normal(key_measurement, xs.shape)
```

```python
xs_cpu = np.array(xs)
for n, row in enumerate(xs):
    # plt.axhline(y=n, c="tab:gray", zorder=0)
    plt.plot((row - row.mean()) / row.std() + n, c="black", zorder=1, linewidth=1.1)
plt.xlabel("State component")
plt.ylabel("State at time $t$")
plt.tight_layout()
plt.savefig("./figures/lorenz96-waves.png", dpi=300)
```

## The ensemble Kalman filter

```python
@chex.dataclass
class EnKFState:
    particles: jax.Array
```

```python
N = 10 ** 4
dt = 0.05
xs = jnp.zeros((N,) + x0.shape)
xs = xs.at[0].set(x0)
xs = rk4(xs, dt, N, f)
ys = xs + jax.random.normal(key_measurement, xs.shape)
```

```python
dt = 0.05
n_particles = 500
```

```python
@partial(jax.vmap, in_axes=(0, None))
def f_next(z, i):
    """
    State function
    """
    return rk4_step(z, 2 * i, dt, f)


def h_next(z, i):
    """
    Measurement function
    """
    keyt = jax.random.fold_in(key_sim, 2 * i + 1)
    keys = jax.random.split(keyt, (n_particles,))
    
    @jax.vmap
    def _next(z, key):
        err = jax.random.normal(key, (D,))
        return z + err
    
    z_next = _next(z, keys)
    return z_next
```

```python
class EnsembleKalmanFilter:
    def __init__(self, state_fn, measurement_fn):
        self.state_fn = state_fn
        self.measurement_fn = measurement_fn
    
    def _predict(self, X, i):
        Xnext = self.state_fn(X, i)
```

```python
def enkf_step(state, y):
    i, X = state

    X_cond = f_next(X, i) # (11)
    Y_cond = h_next(X_cond, i) # (16)

    #Â Compute gain matrix
    matrix_dev = jnp.eye(n_particles) - jnp.ones((n_particles, n_particles)) / n_particles
    X_cond_hat = jnp.einsum("ji,jk->ki", X_cond, matrix_dev)
    Y_cond_hat = jnp.einsum("ji,jk->ki", Y_cond, matrix_dev)

    Mk = jnp.einsum("ji,jk->ik", X_cond_hat, Y_cond_hat) / (n_particles - 1)
    Sk = jnp.einsum("ji,jk->ik", Y_cond_hat, Y_cond_hat) / (n_particles - 1)
    K = jnp.linalg.solve(Sk, Mk)

    X = X_cond + jnp.einsum("ij,kj->ki", K, y - Y_cond)
    
    err_latent = jnp.sqrt(jnp.power(X.mean(axis=0) - xs[i], 2).mean())
    state_next = (i + 1, X)
    
    return state_next, err_latent
```

```python
state_init = (0, X0)
state_final, errs = jax.lax.scan(enkf_step, state_init, ys)
```

## An animation

```python
from matplotlib.animation import FuncAnimation
```

```python
snapshots = 150
N = 60 * snapshots
dt = 0.05
x0 = 8.0 + jax.random.normal(key_init, (D,))
xs = jnp.zeros((N,) + x0.shape)
xs = xs.at[0].set(x0)
xs = rk4(xs, dt, N, f)
ys = xs + jax.random.normal(key_measurement, xs.shape)
xs_cpu = np.array(xs)
```

```python
fig, ax = plt.subplots(figsize=(6.4, 4.8 * 0.7))
def anim(i):
    ax.cla()
    for n, row in enumerate(xs_cpu[i:i + 60][::-1]):
        ax.plot(
            (row - row.mean()) / row.std() + n * 0.4, c="black", linewidth=1.0, zorder=2,
            alpha = 1 - n / 60
        )
    ax.set_ylim(0, 59 * 0.4)
    ax.axis("off")
    plt.tight_layout()

animation = FuncAnimation(fig, anim, frames=snapshots)
animation.save("waves.gif", dpi=130, fps=10)
```
