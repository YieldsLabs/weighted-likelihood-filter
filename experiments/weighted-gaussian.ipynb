```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
%config InlineBackend.figure_format = "retina"
plt.rcParams["font.size"] = 18
```

```python
xrange = np.linspace(-10, 10, 500)
```

```python
dist = norm(loc=0, scale=1)
```

```python
weights = np.linspace(0, 1, 20)[1:]
```

```python
vals = np.exp(dist.logpdf(xrange)[None, :] * weights[:, None] ** 2)
```

```python
colors = plt.cm.viridis(weights)
```

```python
fig, ax = plt.subplots(figsize=(6.4, 4.8 * 1.1))

for val, col in zip(vals, colors):
    ax.plot(xrange, val, c=col)
ax.set_title(r"$\exp(w_t^2\log{\cal N}(x\,\vert\,0, 1))$")


norm = mpl.colors.Normalize(vmin=0, vmax=1)

fig.colorbar(
    mpl.cm.ScalarMappable(norm=norm, cmap="viridis"), ax=ax, orientation="horizontal", shrink=0.8,
    label="$w_t$"
)

plt.xlabel("$x$")
plt.tight_layout()
plt.grid(alpha=0.4)
plt.savefig("./figures/weighted-likelihood-gauss.png", dpi=300);
```
